### Data Visualization

Specifications:
* Figure consists of 1 row by 2 columns
* Figure size is 12 inches x 20 inches
* Amounts must be in thousands (so you'll have to divide by 1000)

Useful links:

```python
import matplotlib.pyplot as plt
```

```python
# Make a 1x3 Figure
%matplotlib inline 

    

# First Graph
# Horizontal Bar Graph
# Bar color: Green ("Green")

# Second Graph
# Vertical Bar
# Bar Color: Orange ("Orange")


# Third Graph


# show plot

plt.show()
```

```python
# Write code here

# Make a 1x3 Figure
%matplotlib inline 


# Write code here
#first graph
fig,axes = plt.subplots(figsize=(12,7))
plt.barh(y=df_country_sorted["Country"].head(10), width=df_country_sorted[2019].head(10)/1000,color='g')
plt.xlabel("Remittances (in Thousands)")
plt.ylabel("Country")
plt.title("Top 10 Sources of Cash Remmitances")
plt.gca().invert_yaxis()

#second graph
fig,axes= plt.subplots(figsize=(12,7))
plt.bar(x=df_continent_sorted["Country"],height=df_continent_sorted[2019]/1000, color="orange")
plt.ylabel("Remmittances (in Thousands)")
plt.xlabel("Continent")
plt.title("Top 10 Sources of Cash Remmitances")

#third graph
fig,axes= plt.subplots(figsize = (12,7))
plt.bar(x=df_USA.index, height = df_USA["remittances"]/1000, color = 'r')
plt.ylabel("Remittances (in Thousands)")
plt.xlabel("Month")
plt.title("Monthly Cash Remittances")


plt.show()
```
